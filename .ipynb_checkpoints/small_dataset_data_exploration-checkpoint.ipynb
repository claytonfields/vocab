```python
import re
import os
import glob
from pathlib import Path
```

```python
def get_text_file(file_path: str):
    with open(file_path, "r") as text_file:

        #read whole file to a string
        data = text_file.read()
        data = data.lower()
        print('File contains', len(data), 'characters\n')

        # words
        words = data.split()
        print('file contains', len(words), 'words\n')
        
        # vocab 
        vocab = set(words)
        print('file contains', len(vocab), 'unique words\n')
    # words: list, vocab: set
    return words, vocab
```

### childes

```python
childes_file = 'aochildes.txt'
childes_path = os.path.join('data', 'childes', childes_file)
childes_words, childes_vocab = get_text_file(childes_path)
```
